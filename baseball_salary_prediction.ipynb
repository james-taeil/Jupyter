```python
# matplorlib 불러오고 한글폰트 설정
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
#matplotlib inline

font_name = font_manager.FontProperties(fname='c:/Windows/Fonts/malgun.ttf').get_name()
rc('font', family=font_name)

picher_features_df = picher[['승','패','세','홀드','블론','경기','선발','이닝','삼진/9',
                             '볼넷/9','홈런/9','BABIP','LOB%','ERA','RA9-WAR','FIP','kFIP','WAR',
                             '연봉(2018)','연봉(2017)']]

def plot_hist_each_column(df):
    plt.rcParams['figure.figsize'] = [20,16]
    fig = plt.figure(1)
    
    for i in range(len(df.columns)):
        ax = fig.add_subplot(5,5,i+1)
        plt.hist(df[df.columns[i]], bins=50)
        ax.set_title(df.columns[i])
    plt.show()
    
plot_hist_each_column(picher_features_df)
```

```python
pd.options.mode.chained_assignment = None

def standard_scaling(df, scale_columns):
    for col in scale_columns:
        series_mean = df[col].mean()
        series_std = df[col].std()
        df[col] = df[col].apply(lambda x: (x-series_mean)/series_std)
    return df

scale_columns = ['승','패','세','홀드','블론','경기','선발','이닝','삼진/9',
                             '볼넷/9','홈런/9','BABIP','LOB%','ERA','RA9-WAR','FIP','kFIP','WAR',
                             '연봉(2017)']
picher_df = standard_scaling(picher, scale_columns)
picher_df = picher_df.rename(columns={'연봉(2018)':'y'})
picher_df.head(5)
```

```python
team_encoding = pd.get_dummies(picher_df['팀명'])
picher_df = picher_df.drop('팀명', axis = 1)
picher_df = picher_df.join(team_encoding)

team_encoding.head(5)
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt

X = picher_df[picher_df.columns.difference(['선수명','y'])]
y = picher_df['y']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state=19)
```

```python
lr = linear_model.LinearRegression()
model = lr.fit(X_train, y_train)

print(lr.coef_)
```

결정계수 (R-squared) or 수정결정계수(ADJ.R-squared) 점수를 봐야된다.

결정계수 점수는 회귀분석이 얼마나 잘 되었는지 평가하는 지표..

1에 가까울수록 데이터를 잘 설명하는 모델이다.

F통계량(F-statistic) 수치를 살펴봐야 된다.

F통계량은 회귀식의 유사성 검정에 사용되는 값으로 F통계량에 대한 P-value인 Prob 수치와 함께 살펴봐야 된다.
0.5 이하면 F통계량이 유의미한 의미를 가진다 라는 결론을 내릴 수있다.

```python
import statsmodels.api as sm
X_train = sm.add_constant(X_train)
model = sm.OLS(y_train,X_train).fit()

model.summary()
```

```python
# matplorlib 불러오고 한글폰트 설정
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
#matplotlib inline
font_path = 'C:/Windows/Fonts/HANBatang.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()

rc('font', family=font_name)
plt.rcParams['figure.figsize'] = [20,16]

#회귀 계수 리스트로 반환
coefs = model.params.tolist()
coefs_series = pd.Series(coefs)

# 변수명을 리스트로 반환
x_labels = model.params.index.tolist()

#회귀 계수 출력
ax = coefs_series.plot(kind='bar')
ax.set_title('feature_coef_graph')
ax.set_xlabel('x_features')
ax.set_ylabel('coef')
ax.set_xticklabels(x_labels)
```

<h3>수정 결정 계수를 sklean의 LinearRegression(선형계수) 클래스로 출력</h3>

```python
X = picher_df[picher_df.columns.difference(['선수명','y'])]
y = picher_df['y']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state=19)

#회귀분석 모델 학습
lr = linear_model.LinearRegression()
model = lr.fit(X_train, y_train)


#회귀분석 모델 평가
print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```

<h3>RMSE값 출력</h3>

이 값이 높으면 예측이 부정확 하다는걸 의미

```python
y_predictions = lr.predict(X_train)
print(sqrt(mean_squared_error(y_train, y_predictions))) #train RMSE 스코어 출력

y_predictions = lr.predict(X_test)
print(sqrt(mean_squared_error(y_test, y_predictions))) #test RMSE 스코어 출력
```

Train 스코어와 Test 스코어의 차이점

어느것이 높아야 할까?

훈련데이터셋은 오로지 모델을 학습 >>> 오차함수를 평가하는데 사용
테스트 데이터셋은 모델에 영향을 미치치 않음 >>> 결과를 예측하기 위한 입력데이터로만 활용

이상적인 결과 >>> 두 값의 차이가 없는 것
값의 차이가 벌어질수록 모델이 훈련 데이터셋에 과적합 된 것

일반적으로 Train스코어가 약간 높다

<h1>다중공선성</h1>

상관관계를 살펴보기 위해 heatmap 방식으로 시각화 사용

1에 가까울수록 높은 연관성

높은 상관관계 일수록 분석에 부정적인 영향을 미치는 것을 의미

```python
import seaborn as sns

# matplorlib 불러오고 한글폰트 설정
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
#matplotlib inline
font_path = 'C:/Windows/Fonts/HANBatang.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()

rc('font', family=font_name)
plt.rcParams['figure.figsize'] = [20,16]

corr = picher_df[scale_columns].corr(method='pearson')
show_cols = ['승','패','세','홀드','블론','경기','선발','이닝','삼진/9',
                             '볼넷/9','홈런/9','BABIP','LOB','ERA','RA9-WAR','FIP','kFIP','WAR',
                             '2017']

plt.rc('font', family='NanumGothicOTF')
sns.set(font_scale=1.5)
hm = sns.heatmap(corr.values,
                cbar=True,
                annot=True,
                square=True,
                fmt='.2f',
                annot_kws={'size':15},
                yticklabels = show_cols,
                xticklabels = show_cols)

plt.tight_layout()
plt.show()
```

다중 공선성은 분석팽창요인 VIF 계수로 평가한다.


일반적으로 10~15를 넘으면 피처는 다중공선성의 문제가 발생했다고 판단

[variance_inflation_factor(x.values, i) for i in range(X.shape[1])] 반복문으로 각 피처마다 함수를 적용해준다


```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['features'] = X.columns
vif.round(1)
```

많은 개수의 피처가 다중공선성 문제를 가짐
