```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

file_path = 'data/도로교통공단_시도 시군구별 교통사고(2019).csv'
traffic_accident = pd.read_csv(file_path) #csv는 콤마로 끊어 잇는데 자동으로 되어있음
print(traffic_accident.info())
traffic_accident.head(10)

traffic_accident.columns
```

```python
traffic_accident = traffic_accident.drop(['시도','시군구'], axis = 1)
print(traffic_accident)
```

```python
traffic_accident.describe()
```

```python
cols = ['사고건수','사망자수','중상자수','경상자수','부상신고자수']
corr = traffic_accident[cols].corr(method='pearson')
print(corr)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt



cols_view = ['사고건수','사망자수','중상자수','경상자수','부상신고자수']
sns.set(font="Malgun Gothic",font_scale=1.5)
plt.figure(figsize=(7,7))
hm = sns.heatmap(corr.values,
                 cbar=True,
                 annot=True,
                 square=True,
                 fmt='.2f',
                 annot_kws={"size":15},
                 yticklabels=cols_view,
                 xticklabels=cols_view)

plt.tight_layout()
plt.show()
```
